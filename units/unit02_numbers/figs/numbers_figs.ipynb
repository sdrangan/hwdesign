## Gaussian distribution

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Wplot = [4, 4, 4]
Fplot = [1, 2, 3]

nplot = len(Wplot)
for i in range(nplot):
    plt.subplot(1, nplot, i+1)
    W = Wplot[i]
    F = Fplot[i]

    # Plot the Gaussian distribution
    x = np.linspace(-4, 4, 1000)
    mean = 0
    std_dev = 1
    gaussian = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)
    plt.plot(x, gaussian, color='blue')

    # Draw the quantization bin edges
    bin_edges = np.arange(-2**(W-1), 2**(W-1)) / (2**F)
    for edge in bin_edges:
        plt.axvline(x=edge, color='red', linestyle='--', alpha=0.5)
    plt.grid()
    plt.yticks([])
    plt.ylim([0, 0.42])
    plt.xlim([-4,4])
    plt.title(f'W={W}, n={F}')

    plt.xlabel('X')

plt.tight_layout()

```
